```python
import cv2
from google.colab.patches import cv2_imshow as cvm
import matplotlib.pyplot as plt
from skimage import io
```

```python
src="https://i.pinimg.com/originals/82/5d/1f/825d1feb6e802a5ff6b93d2bd3c50768.jpg"
```

```python
img=io.imread(src)
```

```python
img
```

```python
cvm(img)
```

```python
img = cv2.resize(img,(500,450))
cvm(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img
```

```python
cvm(gray_img)
```

```python
inverted_img = cv2.bitwise_not(gray_img)
inverted_img
```

```python
cvm(inverted_img)
```

```python
img_smoothing = cv2.GaussianBlur(inverted_img, (25, 15), sigmaX=0, sigmaY=0)
img_smoothing
```

```python
cvm(img_smoothing)
```

```python
def pencil_sketch(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch = pencil_sketch(gray_img, img_smoothing)
cvm(pencil_sketch)
```

**Conclusion** : Thus we have converted the image into pencil sketch using Python in Google Colab.
